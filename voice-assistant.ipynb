```python
import pytesseract
from PIL import Image
from gtts import gTTS
import IPython.display as ipd 
```

```python
language = "en"
```

```python
image = Image.open("/kaggle/input/picture-1/file-page1.jpg")
```

```python
text = pytesseract.image_to_string(image)
```

```python
tts = gTTS(text=text, lang= language, slow=False, tld="com.au")
```

```python
tts.save("output.mp3")
```

```python
ipd.Audio("output.mp3")
```
